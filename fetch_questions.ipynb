```python
import requests
from bs4 import BeautifulSoup

# Step 1: Fetch the web page
url = 'https://leetcode.cn/studyplan/selected-coding-interview/'  # Replace with the actual URL

url = "https://leetcode.cn/studyplan/top-interview-150/"

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/90.0.4430.85 Safari/537.36'
}

response = requests.get(url, headers=headers)

# Check if the request was successful
if response.status_code == 200:
    # Step 2: Parse the HTML
    soup = BeautifulSoup(response.content.decode(), 'html.parser')

    # Step 3: Find all <div> elements with class 'truncate'
    divs = soup.find_all('div', class_='truncate')
    
    # Step 4: Extract the text from each <div>
    texts = [div.get_text(strip=True) for div in divs]

    # Print the texts
    for i, text in enumerate(texts):
        print(f'Text in <div class="truncate">[{i}]: {text}')
else:
    print(f'Failed to retrieve the webpage. Status code: {response.status_code}')
```

```python
import json
json_data = json.loads(soup.find_all("script", type=r"application/json")[0].get_text())
sub_groups = json_data["props"]['pageProps']['dehydratedState']['queries'][0]['state']['data']['studyPlanV2Detail']['planSubGroups']
```

```python
sub_groups[0]['questions'][0]
```

```python
res_dict = {}
for subgroup in sub_groups:
    subgroup_name = subgroup['name']
    res_dict[subgroup_name] = []
    questions = subgroup['questions']
    for question in questions:
        ques_dict = {}
        ques_dict['title'] = question['titleSlug']
        ques_dict['id'] = question['id']
        ques_dict['translatedTitle'] = question['translatedTitle']
        ques_dict['tag'] = ''
        tags = question['topicTags']
        # for tag in tags:
        for i in range(len(tags)):
            tag = tags[i]
            if i == len(tags) - 1:
                ques_dict['tag'] += tag['name']
            else:
                ques_dict['tag'] += tag['name'] + ", "
        res_dict[subgroup_name].append(ques_dict)
```

```python
with open('question_list.md', 'w') as f:
    for key in res_dict.keys():
        f.write(f"## {key}\n\n")
        for question_dict in res_dict[key]:
            f.write(f"### {question_dict['translatedTitle']}\n\n- id: {question_dict['id']}\n- name: {question_dict['title']}\n- tag: {question_dict['tag']}\n\n")
        f.write('\n--- \n\n\n')
```
